### Grammar Corrector (Normal Transformer) Model

```python
model_name = 'deep-learning-analytics/GrammarCorrector'
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to(torch_device)
```

```python
def correct_grammar(input_text,num_return_sequences):
    batch = tokenizer([input_text],truncation=True,padding='max_length',max_length=64, return_tensors="pt").to(torch_device)
    translated = model.generate(**batch,max_length=64,num_beams=2, num_return_sequences=num_return_sequences, temperature=1.5)
    tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
    return tgt_text
```

```python
#Running grammar corrector using normal transformer on each input grammatically incorrect sentence
list_exp = expressions['Expression'].tolist()
list_corrected_exp_3 = []
for i in list_exp:
    text = i
    output = correct_grammar(text,num_return_sequences=1)
    print('Corrected_output: ' + str(output[0]))
    list_corrected_exp_3.append(output)
```

```python
#Saving the results in a csv file
pd.DataFrame(list_corrected_exp_3).to_csv('list_corrected_exp_GrammarCorrector.csv')
```

### Grammar Correction (Wave2Vec2 Grammar Fixer) Model

```python
model_name = "flexudy/t5-small-wav2vec2-grammar-fixer"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
sent = """GOING ALONG SLUSHY COUNTRY ROADS AND SPEAKING TO DAMP AUDIENCES IN DRAUGHTY SCHOOL ROOMS DAY AFTER DAY FOR A FORTNIGHT HE'LL HAVE TO PUT IN AN APPEARANCE AT SOME PLACE OF WORSHIP ON SUNDAY MORNING AND HE CAN COME TO US IMMEDIATELY AFTERWARDS"""

input_text = "fix: { " + sent + " } </s>"

input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=256, truncation=True, add_special_tokens=True)

outputs = model.generate(
    input_ids=input_ids,
    max_length=256,
    num_beams=4,
    repetition_penalty=1.0,
    length_penalty=1.0,
    early_stopping=True
)

sentence = tokenizer.decode(outputs[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

print(f"{sentence}")
```

```python
#Running grammar corrector using wave2vec2 grammar fixer model on each input grammatically incorrect sentence
list_exp = expressions['Expression'][0:3000].tolist()
list_corrected_exp_3 = []
for i in list_exp:
    text = i
    input_text = "fix: { " + text + " } </s>"

    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=256, truncation=True, add_special_tokens=True)

    outputs = model.generate(
        input_ids=input_ids,
        max_length=256,
        num_beams=4,
        repetition_penalty=1.0,
        length_penalty=1.0,
        early_stopping=True
    )

    sentence = tokenizer.decode(outputs[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

    print('Corrected_output: ' + str(sentence))
    list_corrected_exp_3.append(sentence)
```

```python
#Saving the results in a csv file
pd.DataFrame(list_corrected_exp_3).to_csv('list_corrected_exp_t5-small-wav2vec2-grammar-fixer.csv')
```

### Grammar Error Corrector V1 (Prithviraj)

```python
model_name = "prithivida/grammar_error_correcter_v1"
torch_device = 'cuda' if torch.cuda.is_available() else 'cpu'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to(torch_device)
```

```python
def correct_grammar(input_text,num_return_sequences):
    batch = tokenizer([input_text],truncation=True,padding='max_length',max_length=64, return_tensors="pt").to(torch_device)
    translated = model.generate(**batch,max_length=64,num_beams=2, num_return_sequences=num_return_sequences, temperature=1.5)
    tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
    return tgt_text
```

```python
list_exp = expressions['Expression'].tolist()
list_corrected_exp_3 = []
for i in list_exp:
    text = i
    output = correct_grammar(text,num_return_sequences=1)
    print('Corrected_output: ' + str(output[0]))
    list_corrected_exp_3.append(output)
```

```python
#Saving the results in a csv file
pd.DataFrame(list_corrected_exp_3).to_csv('list_corrected_exp_grammar_error_correcter_v1.csv')
```

```python
import pandas as pd
from happytransformer import HappyTextToText,TTSettings
```

```python
text = "gec: " + "We no open, sorrry"
happy_tt = HappyTextToText("T5",  "prithivida/grammar_error_correcter_v1")
settings = TTSettings(do_sample=True, top_k=10, temperature=0.5,  min_length=1, max_length=100)
result = happy_tt.generate_text(text, args=settings)
print(result.text)
```

```python
data=pd.read_csv('client_result_2043_with_partial_human_QAs.csv')
```

```python
def grammer_pipeline(text):
    text = "gec: " + text
    result = happy_tt.generate_text(text, args=settings)
    
    return result.text

data['Result'] =  data['Expression'].apply(grammer_pipeline)
```

```python
data.to_csv('Results_grammar_corrector_w_happy_transf.csv')
```
